```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data=pd.read_csv('movie_dataset.csv')
```

```python
features=['keywords','cast','genres','director']
cast=['cast']
director=['director']
genres=['genres']
```

```python
def combine_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
for i in features:
    data[i] = data[i].fillna(' ')   
for i in cast:
    data[i] = data[i].fillna(' ') 
for i in director:
    data[i] = data[i].fillna(' ') 
for i in genres:
    data[i] = data[i].fillna(' ') 
```

```python
data["combined_features"] = data.apply(combine_features,axis=1)
data.columns
data['title']=data['title'].str.lower()
data.head()
```

```python
cv=CountVectorizer()
```

```python
count_mat1=cv.fit_transform(data['combined_features'])
count_mat2=cv.fit_transform(data['cast'])
count_mat3=cv.fit_transform(data['director'])
count_mat4=cv.fit_transform(data['genres'])
```

```python
cosine_sim1 = cosine_similarity(count_mat1)
cosine_sim2 = cosine_similarity(count_mat2)
cosine_sim3 = cosine_similarity(count_mat3)
cosine_sim4 = cosine_similarity(count_mat4)
```
